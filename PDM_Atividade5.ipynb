### **Desafio 1: Sistema de Gerenciamento de Consultas Médicas (SaaS)**

#### **Descrição do Projeto**
Desenvolver um SaaS para gerenciamento de consultas médicas. O sistema deve ser voltado para médicos e pacientes, utilizando o padrão arquitetural **MVVM** com gerenciamento de estado via **Provider**, e integração com **Hasura GraphQL** para backend.

---

#### **Funcionalidades Obrigatórias**
1. **Cadastro e Login**:
   - **Login com Google** (via Firebase).
   - Cadastro de médicos e pacientes (diferenciando os papéis no sistema).

2. **Gerenciamento de Consultas**:
   - O médico pode:
     - Criar, editar e cancelar consultas.
     - Ver a lista de pacientes cadastrados.
   - O paciente pode:
     - Agendar uma consulta com base na disponibilidade do médico.
     - Visualizar o histórico de consultas.

3. **Notificações em Tempo Real**:
   - Utilizar **Hasura Subscriptions** para que as mudanças no banco de dados (consultas agendadas ou canceladas) sejam refletidas em tempo real no aplicativo.

4. **Painéis Informativos**:
   - **Médico**: Visualizar gráficos e estatísticas sobre consultas (número de pacientes atendidos, horários mais agendados, etc.).
   - **Paciente**: Visualizar histórico e próximos compromissos.

5. **Upload de Arquivos**:
   - O paciente pode anexar documentos (ex.: resultados de exames) às consultas.
   - Utilizar **Firebase Storage** para armazenar os arquivos.

6. **Documentação e Apresentação**:
   - Criar um README detalhado no GitHub explicando a arquitetura do projeto, instalação e configuração do Hasura, Firebase, etc.
   - Vídeo explicativo sobre o sistema.

---

#### **Tecnologias Recomendadas**
- **Frontend**: Flutter (MVVM com Provider).
- **Backend**: Hasura (GraphQL para consultas e mutações).
- **Autenticação**: Firebase Authentication (login com Google).
- **Banco de Dados**: PostgreSQL via Hasura.
- **Armazenamento de Arquivos**: Firebase Storage.
- **Gráficos**: Pacote `fl_chart` para Flutter.

---

#### **Critérios de Avaliação**
- **Complexidade Técnica**: Utilização correta do padrão MVVM e integração com GraphQL.
- **Interface do Usuário**: Design simples, responsivo e intuitivo.
- **Documentação Completa**: Passos claros para configurar e rodar o projeto.
- **Entrega no GitHub**: Código organizado, bem comentado, com commits significativos.

### **Desafio 2: Sistema de Gerenciamento de Projetos (SaaS)**

#### **Descrição do Projeto**
Criar um SaaS para gerenciamento de projetos e tarefas em equipe, focado em times pequenos. O sistema deve permitir o gerenciamento completo do fluxo de trabalho.

---

#### **Funcionalidades Obrigatórias**
1. **Autenticação**:
   - Login e cadastro via **Google** (Firebase Authentication).
   - Diferenciação de papéis: Administrador e Usuários.

2. **Gerenciamento de Projetos e Tarefas**:
   - **Administrador**:
     - Criar, editar e excluir projetos.
     - Atribuir tarefas para os usuários da equipe.
   - **Usuários**:
     - Visualizar tarefas atribuídas e marcar como concluídas.

3. **Quadro Kanban**:
   - Implementar um quadro Kanban para organizar as tarefas dentro dos projetos (ex.: “A Fazer”, “Em Andamento”, “Concluído”).

4. **Integração com Hasura GraphQL**:
   - Gerenciar projetos, tarefas e usuários no banco de dados PostgreSQL via Hasura.
   - Utilizar **Hasura Subscriptions** para atualizações em tempo real no quadro Kanban.

5. **Notificações e Comentários**:
   - Enviar notificações em tempo real para os membros da equipe quando uma tarefa for criada ou alterada.
   - Permitir que os usuários comentem nas tarefas (como um chat interno).

6. **Dashboard**:
   - Painéis para o administrador visualizarem:
     - Progresso dos projetos (percentual de tarefas concluídas).
     - Estatísticas do time (quantidade de tarefas por status, membros mais ativos, etc.).

7. **Documentação e Apresentação**:
   - README detalhado no GitHub explicando o sistema e a arquitetura.
   - Vídeo explicando o funcionamento e os recursos.

---

#### **Tecnologias Recomendadas**
- **Frontend**: Flutter com MVVM e Provider.
- **Backend**: Hasura (GraphQL).
- **Banco de Dados**: PostgreSQL.
- **Autenticação**: Firebase Authentication (login com Google).
- **Armazenamento**: Firebase Storage (caso necessário upload de arquivos).
- **UI**: Uso de pacotes como `drag_and_drop_lists` para o quadro Kanban e `fl_chart` para gráficos.

---

#### **Critérios de Avaliação**
- **Implementação Técnica**: Uso eficiente de MVVM e controle de estado.
- **Design**: Interface amigável e bem estruturada.
- **Funcionalidade Completa**: Todas as funções descritas funcionando corretamente.
- **Documentação e Vídeo**: Entrega no GitHub com explicação detalhada.

### **Desafio 3: Sistema de Gestão de Fazenda (SaaS) para Pequenos e Médios Produtores**  

#### **Descrição do Projeto**  
Desenvolver um **SaaS para gerenciamento de fazendas**, ajudando pequenos e médios produtores a controlar suas operações agrícolas e pecuárias. O sistema deve permitir o **cadastro e acompanhamento de cultivos, controle de estoque, rastreamento financeiro e gerenciamento de rebanhos**. O aplicativo deve ser construído utilizando **Flutter** com o padrão **MVVM**, gerenciamento de estado com **Provider**, backend utilizando **Hasura GraphQL** e autenticação via **Google**.  

---

#### **Funcionalidades Obrigatórias**  

### **1. Autenticação e Acesso**  
- **Login com Google** (Firebase Authentication).  
- Diferenciação de **papéis**: **Produtor** (dono da fazenda) e **Funcionários** (acesso restrito).  

### **2. Cadastro e Gerenciamento da Fazenda**  
- O produtor pode cadastrar informações sobre a **fazenda**, como:  
  - Nome da fazenda, localização e tamanho da área cultivável.  
  - Tipos de cultivo ou criação de animais.  

### **3. Controle de Produção Agrícola**  
- Cadastro e acompanhamento de **culturas** (exemplo: soja, milho, café).  
- Registro de **safras** e **previsão de colheita**.  
- Monitoramento de **insumos utilizados** e **custo por hectare**.  

### **4. Gerenciamento de Pecuária**  
- Cadastro de **rebanhos** e animais individuais (exemplo: bovinos, suínos, aves).  
- Registro de **vacinação e alimentação** por animal ou grupo.  
- Controle de **produção de leite ou ovos**, caso aplicável.  

### **5. Controle Financeiro**  
- **Lançamento de despesas e receitas**, categorizando por tipo (exemplo: fertilizantes, ração, venda de soja, etc.).  
- **Gráficos financeiros** mostrando faturamento e custos da fazenda.  
- Cálculo de **lucro/prejuízo por safra ou período**.  

### **6. Estoque e Insumos**  
- Registro e controle do **estoque de insumos** (exemplo: sementes, fertilizantes, defensivos, ração).  
- Notificações para **baixa no estoque**.  

### **7. Relatórios e Painéis**  
- Painel para visualizar **dados gerais** da fazenda.  
- Relatórios por **safra, rebanho e financeiro**.  
- Gráficos sobre produção e rentabilidade.  

---

#### **Funcionalidades Extras (Para Alunos Que Querem Desafios Avançados 🚀)**  
🔹 **Previsão de produção usando Machine Learning** com base nos dados cadastrados.  
🔹 **Notificações em tempo real** (exemplo: lembrete de vacinação de animais).  
🔹 **Integração com API Meteorológica** para exibir previsões do tempo.  
🔹 **Modo offline** com sincronização dos dados quando a conexão voltar.  

---

### **Tecnologias Recomendadas**  
- **Frontend**: Flutter com MVVM e Provider.  
- **Backend**: Hasura GraphQL.  
- **Autenticação**: Firebase Authentication (Login com Google).  
- **Banco de Dados**: PostgreSQL via Hasura.  
- **Armazenamento de Arquivos**: Firebase Storage (para fotos de animais, documentos da fazenda, etc.).  
- **Gráficos e Relatórios**: Pacote `fl_chart` para Flutter.
